## 標準化

```python
A = np.linspace(0, 1, 100).reshape(-1, 1) + np.random.randint(1, 10, size=100).reshape(-1, 1)
B = np.linspace(10000, 100000, 100).reshape(-1, 1) + np.random.randint(1, 300000, size=100).reshape(-1, 1)

ss = StandardScaler()
ss.fit(A)
C = ss.transform(A)
D = ss.fit_transform(B)

df = pd.DataFrame({
    "A": A.flatten(),
    "B": B.flatten(),
    "C": C.flatten(),
    "D": D.flatten()
})

fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax1.hist(A)
ax2 = fig.add_subplot(2, 2, 2)
ax2.hist(C)
ax3 = fig.add_subplot(2, 2, 3)
ax3.hist(B)
ax4 = fig.add_subplot(2, 2, 4)
ax4.hist(D)
```
